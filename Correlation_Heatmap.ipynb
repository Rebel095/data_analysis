#Как создать тепловую карту корреляции

Давайте разберемся, как можно использовать тепловые карты на практике. Один из интересных примеров — это корреляционная тепловая карта. Корреляционная матрица — это полезный инструмент, который помогает анализировать данные и находить связи между ними. А тепловая карта просто делает эту информацию более наглядной и удобной для восприятия, представляя ее в виде цветной схемы.

## Создание тестового датафрейма

```python
# импорт библиотек
import pandas as pd
import random
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = []
# создадим датафрейм из 50-ти строк
for i in range(50):
  # метод randint() возвращает целое число выбранного элемента из указанного диапазона (включая оба).
  a = random.randint(1, 100)
  b = random.randint(1, 100)
  c = random.randint(1, 100)
  d = random.randint(1, 100)
  e = random.randint(1, 100)
  f = random.randint(1, 100)
  g = random.randint(1, 100)
  h = random.randint(1, 100)
  x = {'A': a, 'B': b, 'C': c, 'D': d, 'E': e, 'F': f, 'G': g, 'H': h}
  data.append(x)

df = pd.DataFrame(data)
```

```python
df
```

## Создаем матрицу корреляции

```python
# Создание матрицы корреляции
correlation_matrix = df.corr() # метод .corr() вычисляет корреляцию между всеми числовыми столбцами датафрейма.
correlation_matrix
```

Результат — это матрица корреляции, где значения показывают, насколько сильно связаны между собой столбцы.

## Создание тепловой карты

### Простая Seaborn Heatmap

```python
sns.heatmap(correlation_matrix)
```

Базовая тепловая карта выглядит не такой наглядной, давайте её улучшим.

Взгляните на [список аргументов тепловой карты Seaborn](https://seaborn.pydata.org/generated/seaborn.heatmap.html)

`seaborn.heatmap(data, *, vmin=None, vmax=None, cmap=None, center=None, robust=False, annot=None, fmt='.2g', annot_kws=None, linewidths=0, linecolor='white', cbar=True, cbar_kws=None, cbar_ax=None, square=False, xticklabels='auto', yticklabels='auto', mask=None, ax=None, **kwargs)`

где

`vmin, vmax` — задают диапазон значений, которые служат основой для цветовой карты


`cmap` — задают конкретную цветовую палитру, которую мы хотим использовать (ознакомьтесь с библиотекой огромного диапазона [цветовых палитр](https://python-graph-gallery.com/92-control-color-in-seaborn-heatmaps/))

`center` — принимает число с плавающей точкой для центрирования цветовой карты; если `cmap` не указан, изменяет цвета в цветовой карте по умолчанию; если установлено значение `True` — изменяет все цвета цветовой карты на синие

`annot` — если установлено значение `True`, значения корреляции становятся видимыми на цветных ячейках

`cbar` — если установлено значение `False`, цветовая полоса (которая служит легендой) исчезает

```python
# увеличьте размер тепловой карты
plt.figure(figsize=(16, 6))

# cохраните объект тепловой карты в переменной, чтобы легко получить к нему доступ, когда вы захотите включить больше функций (например, заголовок).
# Установите диапазон значений, которые будут отображаться на цветовой карте, от -1 до 1,
# и установите аннотацию на True, чтобы отобразить значения корреляции на тепловой карте
heatmap = sns.heatmap(correlation_matrix, vmin=-1, vmax=1, annot=True)

# дайте заголовок тепловой карте. Pad определяет расстояние заголовка от верхней части тепловой карты.
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':12}, pad=12);
```

**Расходящаяся цветовая палитра** , которая имеет заметно отличающиеся цвета на двух концах диапазона значений с бледной, почти бесцветной средней точкой, работает гораздо лучше с корреляционными тепловыми картами, чем цветовая карта по умолчанию.

```python
plt.figure(figsize=(16, 6))
heatmap = sns.heatmap(correlation_matrix.corr(), vmin=-1, vmax=1, annot=True, cmap='BrBG')
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':18}, pad=12);

# сохранить тепловую карту как файл .png
# dpi - устанавливает разрешение сохраняемого изображения в точках/дюймах
# bbox_inches - если установлено значение «tight» - не позволяет обрезать метки
plt.savefig('heatmap.png', dpi=300, bbox_inches='tight')
```

Более сильная корреляция на обоих концах спектра проявляется в более темных тонах, а более слабая корреляция — в более светлых.

## Треугольная тепловая карта корреляции

Взгляните на любую из тепловых карт корреляции выше. Если вы отрежете половину по диагональной линии, отмеченной 1-цами, вы не потеряете никакой информации. Давайте тогда разрежем тепловую карту пополам и оставим только нижний треугольник.

Аргумент `mask` тепловой карты Siaborn оказывается полезным, когда мы хотим охватить часть тепловой карты.

`mask` — принимает в качестве аргумента логический массив или фрейм данных; если определено, ячейки становятся невидимыми для значений, где маска имеет значение `True`

Давайте используем функцию numpy `np.triu()`, чтобы изолировать верхний треугольник матрицы, одновременно превратив все значения в нижнем треугольнике в 0. (Функция `np.tril()` сделает то же самое, только для нижнего треугольника.)

Использование функции `np.ones_like()` изменит все остальные значения на 1.

```python
np.triu(np.ones_like(correlation_matrix.corr()))
```

```python
plt.figure(figsize=(16, 6))
# определить маску, чтобы установить значения в верхнем треугольнике на True
mask = np.triu(np.ones_like(correlation_matrix.corr()))
heatmap = sns.heatmap(correlation_matrix.corr(), mask=mask, vmin=-1, vmax=1, annot=True, cmap='BrBG')
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':18}, pad=16);
```

## Корреляция независимых переменных с зависимой переменной

Однако часто мы хотим создать цветную карту, показывающую силу корреляции между каждой независимой переменной, которую мы хотим включить в нашу модель, и зависимой переменной.

Допустим мы хотим посмотреть как коррелирую все остальные признаки с нашим `А` признаком.

```python
correlation_matrix.corr()[['A']].sort_values(by='A', ascending=False)
```

Давайте используем эти данные в нашей тепловой карте.

```python
plt.figure(figsize=(8, 5))
heatmap = sns.heatmap(correlation_matrix.corr()[['A']].sort_values(by='A', ascending=False), vmin=-1, vmax=1, annot=True, cmap='BrBG')
heatmap.set_title('Correlation with variable A', fontdict={'fontsize':18}, pad=16);
```
